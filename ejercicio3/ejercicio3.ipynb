```python
# Cargar el dataset
file_path = '/content/drive/My Drive/primerParcial354/adult.csv'
data = pd.read_csv(file_path)

# Verificar los nombres de las columnas
print(data.columns)

import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer


# Ajustar los nombres de las columnas según sea necesario
categorical_columns = ['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country']
numerical_columns = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']

# 1. Imputación de valores faltantes
# Imputación para variables categóricas (con la moda)
categorical_imputer = SimpleImputer(strategy='most_frequent')
data[categorical_columns] = categorical_imputer.fit_transform(data[categorical_columns])

# Imputación para variables numéricas (con la mediana)
numerical_imputer = SimpleImputer(strategy='median')
data[numerical_columns] = numerical_imputer.fit_transform(data[numerical_columns])


```

## Pre procesamiento 2 OneHotEncoder para variables categóricas

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer


# Seleccionar columnas categóricas
categorical_columns = ['sex']

# 2. OneHotEncoder para la columna categórica 'sex'
encoder = OneHotEncoder(drop='first', sparse=False)
encoded_data = pd.DataFrame(encoder.fit_transform(data[categorical_columns]))
encoded_data.columns = encoder.get_feature_names_out(categorical_columns)

# Eliminar la columna original 'sex' y agregar las columnas codificadas
data = data.drop(columns=categorical_columns)
data = pd.concat([data.reset_index(drop=True), encoded_data.reset_index(drop=True)], axis=1)

```

# Pre procesamiento 3 Escalado

```python
from sklearn.preprocessing import StandardScaler

# Seleccionar columnas numéricas
numerical_columns = ['education_num']

# Aplicar escalado
scaler = StandardScaler()
scaled_data = pd.DataFrame(scaler.fit_transform(data[numerical_columns]), columns=numerical_columns)

# Combinar datos escalados con el resto de los datos
data[numerical_columns] = scaled_data
```

# Pre procesamiento 4 Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Aplicar LabelEncoder
label_encoder = LabelEncoder()
data['income'] = label_encoder.fit_transform(data['income'])
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Imputación de valores faltantes si es necesario
# Para simplificar, asumimos que no hay valores faltantes en 'education'
# Si hubiera valores faltantes, se podrían imputar como sigue:
categorical_columns = ['education']
imputer = SimpleImputer(strategy='most_frequent')
data[categorical_columns] = imputer.fit_transform(data[categorical_columns])

# Definir el orden para la columna 'education' basado en los valores únicos presentes en los datos
# Crear el objeto OrdinalEncoder
encoder = OrdinalEncoder()

race_data = data[['race']]
# Transformar los datos
race_encoded = encoder.fit_transform(race_data)

data['race'] = race_encoded

# Mostrar los valores codificados y las categorías
print("Datos codificados:\n", race_encoded[:10])
print("Categorías:", encoder.categories_)

# Guardar el dataset preprocesado
data.to_csv('/content/drive/My Drive/primerParcial354/adult_preprocesado.csv', index=False)
```
