```python
# Definir los nodos del grafo
nodos = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']



# Función para generar todas las permutaciones usando el algoritmo de Heap
def generar_permutaciones(nodos):
    n = len(nodos)
    c = [0] * n
    rutas_posibles = set()
    x = ''.join(nodos)
    # Añadir la permutación inicial
    rutas_posibles.add(x)
    i = 0
    while i < n:
        if c[i] < i:
            if i % 2 == 0:
                nodos[0], nodos[i] = nodos[i], nodos[0]
            else:
                nodos[c[i]], nodos[i] = nodos[i], nodos[c[i]]
            y = ''.join(nodos)
            rutas_posibles.add(y)
            c[i] += 1
            i = 0
        else:
            c[i] = 0
            i += 1

    return rutas_posibles


sin_repetidos = generar_permutaciones(nodos)

# Generar todas las rutas posibles
# Mostrar algunas de las rutas generadas
print(len(sin_repetidos))
print(sin_repetidos)

```
